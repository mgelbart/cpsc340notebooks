## t-SNE demo

```python
from sklearn.datasets import load_digits

data = load_digits()
```

```python
i = 10
plt.imshow(data['images'][i],cmap='gray');
plt.title("This is a %d" % data['target'][i]);
```

```python
X = data['data']
y = data['target']
```

```python
tsne = TSNE()
Z = tsne.fit_transform(X)
```

```python
plt.scatter(Z[:,0], Z[:,1], alpha=0.3);
plt.xlabel('$z_1$');
plt.ylabel('$z_2$');
```

```python
plt.scatter(Z[:,0], Z[:,1], alpha=0.3, c=y);
plt.xlabel('$z_1$');
plt.ylabel('$z_2$');
```
